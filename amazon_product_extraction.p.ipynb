```python
from bs4 import BeautifulSoup
import pandas as pd

# Load the HTML content
file_path = "C:\\Users\\user\\Dropbox\\PC\\Downloads\\amazon_23.html"
with open(file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')

# Define lists to store the extracted data
product_names = []
product_prices = []
product_ratings = []
rating_counts = []

# Extract the required data
items = soup.find_all('div', {'data-component-type': 's-search-result'})

for item in items:
    # Product Name
    name_tag = item.find('h2', class_='a-size-mini a-spacing-none a-color-base s-line-clamp-4')
    if name_tag:
        product_names.append(name_tag.get_text(strip=True))
    else:
        product_names.append(None)
    
    # Product Price
    price_tag = item.find('span', class_='a-offscreen')
    if price_tag:
        product_prices.append(price_tag.get_text(strip=True))
    else:
        product_prices.append(None)
    
    # Product Rating
    rating_tag = item.find('span', class_='a-icon-alt')
    if rating_tag:
        product_ratings.append(rating_tag.get_text(strip=True))
    else:
        product_ratings.append(None)
    
    # Rating Count
    rating_count_tag = item.find('span', {'aria-label': True, 'class': None})
    if rating_count_tag:
        rating_counts.append(rating_count_tag.get_text(strip=True))
    else:
        rating_counts.append(None)

# Create a DataFrame
data = {
    'Product Name': product_names,
    'Price': product_prices,
    'Rating': product_ratings,
    'Rating Count': rating_counts
}

df = pd.DataFrame(data)

# Save the DataFrame to a CSV file with the new path
output_file = "C:\\Users\\user\\Dropbox\\PC\\Documents\\amazon_combined_products.csv"
df.to_csv(output_file, index=False)

# Display the DataFrame to the user
import ace_tools as tools
tools.display_dataframe_to_user(name="Amazon Products Data", dataframe=df)

output_file
```
